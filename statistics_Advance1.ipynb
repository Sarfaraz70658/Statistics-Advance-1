```python

# Q-2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define a discrete uniform distribution (e.g., integers between 1 and 6 for a die)
low, high = 1, 6  # Range: inclusive of `low`, inclusive of `high`
distribution = randint(low, high + 1)  # Use randint to define the uniform distribution

# Get the range of values
x = np.arange(low, high + 1)

# Calculate the PMF (probability mass function)
pmf = distribution.pmf(x)

# Plot the PMF
plt.bar(x, pmf, color="skyblue", alpha=0.7, edgecolor="black")
plt.title("PMF of a Discrete Uniform Distribution (Die Roll)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.xticks(x)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

```

```python

# Q 3- Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    """
    Calculate the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability of observing x in the Bernoulli distribution.
    """
    if x not in [0, 1]:
        raise ValueError("Invalid outcome. x must be 0 (failure) or 1 (success).")
    if not (0 <= p <= 1):
        raise ValueError("Invalid probability. p must be between 0 and 1.")

    return p if x == 1 else (1 - p)

# Example usage:
p_success = 0.7  # Probability of success
outcome = 1  # Observing success
pdf_value = bernoulli_pdf(outcome, p_success)
print(f"The PDF value for outcome {outcome} with success probability {p_success} is {pdf_value}")
```

```python

# Q-9  Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a given dataset and plot their histogram.

    Parameters:
    data (list or np.array): The dataset to calculate Z-scores for.

    Returns:
    np.array: An array of calculated Z-scores.
    """
    # Convert data to a NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero. Cannot calculate Z-scores.")

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot the histogram of Z-scores
    plt.hist(z_scores, bins=20, color='skyblue', alpha=0.7, edgecolor='black', density=True)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    return z_scores

# Example usage:
# Example dataset
data = [10, 20, 30, 40, 50, 60, 70]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python

# Q-12 Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).

import numpy as np
import matplotlib.pyplot as plt

def plot_standard_normal_distribution():
    """
    Calculate and plot the Standard Normal Distribution (mean = 0, std = 1).
    """
    # Generate values for the x-axis (range: -4 to 4, as most data lies within this range for normal distributions)
    x = np.linspace(-4, 4, 500)

    # Calculate the standard normal distribution (probability density function, PDF)
    y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

    # Plot the standard normal distribution
    plt.plot(x, y, color="blue", label="Standard Normal Distribution")
    plt.title("Standard Normal Distribution (mean=0, std=1)")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.grid(linestyle="--", alpha=0.7)
    plt.legend()
    plt.show()

# Example usage
plot_standard_normal_distribution()
```

```python

# Q-14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import scipy.stats as stats

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    data_point (float): The value for which the Z-score is to be calculated.
    mean (float): The mean of the dataset.
    std_dev (float): The standard deviation of the dataset.

    Returns:
    float: The calculated Z-score.
    """
    if std_dev == 0:
        raise ValueError("Standard deviation cannot be zero.")

    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    """
    Compare the Z-score to a standard normal distribution.

    Parameters:
    z_score (float): The calculated Z-score.

    Returns:
    tuple: The probability below and above the Z-score in the standard normal distribution.
    """
    # Cumulative probability below the Z-score
    prob_below = stats.norm.cdf(z_score)
    # Probability above the Z-score
    prob_above = 1 - prob_below
    return prob_below, prob_above

# Example usage
data_point = 75
mean = 60
std_dev = 10

# Calculate Z-score
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point}: {z_score:.2f}")

# Compare Z-score to the standard normal distribution
prob_below, prob_above = compare_to_standard_normal(z_score)
print(f"Probability below Z-score: {prob_below:.4f}")
print(f"Probability above Z-score: {prob_above:.4f}")
```

```python

# Q-15  Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def hypothesis_testing(sample, population_mean, population_std, significance_level=0.05):
    """
    Perform hypothesis testing using Z-statistics.

    Parameters:
    sample (list or np.array): The sample data.
    population_mean (float): The mean of the population.
    population_std (float): The standard deviation of the population.
    significance_level (float): The significance level (default is 0.05).

    Returns:
    str: Conclusion of the hypothesis test.
    """
    # Convert the sample to a NumPy array
    sample = np.array(sample)

    # Calculate the sample mean and size
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the critical value (two-tailed test)
    critical_value = norm.ppf(1 - significance_level / 2)

    # Compare the Z-statistic to the critical value
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"Critical Value: ±{critical_value:.4f}")

    if abs(z_statistic) > critical_value:
        return "Reject the null hypothesis (H0). The sample mean significantly differs from the population mean."
    else:
        return "Fail to reject the null hypothesis (H0). There is no significant difference between the sample mean and the population mean."

# Example Usage
sample_data = [12, 15, 14, 10, 13, 14, 16, 11, 15, 13]
population_mean = 12
population_std = 3
significance_level = 0.05

result = hypothesis_testing(sample_data, population_mean, population_std, significance_level)
print(result)
```

```python

# Q-16  Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import t

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
    data (list or np.array): The dataset.
    confidence (float): The confidence level (default is 95%).

    Returns:
    tuple: Lower bound and upper bound of the confidence interval.
    """
    # Convert data to a NumPy array
    data = np.array(data)

    # Calculate sample mean and standard error
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(len(data))  # Use sample standard deviation

    # Determine the critical value (t-distribution for smaller samples, z-distribution for large samples)
    critical_value = t.ppf((1 + confidence) / 2, df=len(data) - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * std_err

    # Calculate the confidence interval
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
data = [15, 18, 21, 24, 27, 30, 33, 36, 39, 42]
confidence_level = 0.95

lower, upper = calculate_confidence_interval(data, confidence=confidence_level)
print(f"A {confidence_level * 100}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

# Interpretation
mean = np.mean(data)
print(f"This means that we are {confidence_level * 100}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")

```

```python

# Q-17  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy.stats import t

def generate_normal_data(mean, std_dev, size):
    """
    Generate data from a normal distribution.
    """
    return np.random.normal(mean, std_dev, size)

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.
    """
    # Convert data to a NumPy array
    data = np.array(data)

    # Calculate sample mean and standard error
    sample_mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(len(data))  # Use sample standard deviation

    # Determine critical value for t-distribution
    critical_value = t.ppf((1 + confidence) / 2, df=len(data) - 1)

    # Calculate margin of error
    margin_of_error = critical_value * std_err

    # Compute the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return sample_mean, lower_bound, upper_bound

# Example usage
mean, std_dev, size = 50, 10, 100  # Normal distribution parameters
data = generate_normal_data(mean, std_dev, size)

# Confidence level
confidence_level = 0.95

# Calculate the confidence interval
sample_mean, lower, upper = calculate_confidence_interval(data, confidence=confidence_level)
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{confidence_level * 100}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

# Interpretation
print(f"This means we are {confidence_level * 100}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

```python

# Q-18  Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution

# Generate values for the x-axis
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 500)  # Range of x values (±4σ)

# Calculate the PDF for the normal distribution
pdf = norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.plot(x, pdf, color="blue", label=f"Normal Distribution (mean={mean}, std={std_dev})")
plt.title("Probability Density Function of a Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(linestyle="--", alpha=0.7)
plt.legend()
plt.show()
```

```python

# Q-19  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
from scipy.stats import poisson

def calculate_poisson_cdf(lam, k):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.

    Parameters:
    lam (float): The mean (λ) of the Poisson distribution (average number of events).
    k (int): The number of events to evaluate the CDF at.

    Returns:
    float: The CDF value.
    """
    # Calculate the CDF
    cdf_value = poisson.cdf(k, lam)
    return cdf_value

# Example usage
lambda_value = 5  # Average number of events (e.g., 5 events per interval)
k_value = 3       # Evaluate the CDF up to 3 events

cdf_result = calculate_poisson_cdf(lambda_value, k_value)
print(f"The probability of observing up to {k_value} events when the average is {lambda_value}: {cdf_result:.4f}")

# Interpretation
print(f"This means there is a {cdf_result * 100:.2f}% chance that the number of events will be 3 or fewer.")
```

```python

# Q-20  Simulate a random variable using a continuous uniform distribution and calculate its expected value.

import numpy as np

def simulate_uniform_random_variable(low, high, size):
    """
    Simulate a continuous uniform random variable and calculate its expected value.

    Parameters:
    low (float): The lower bound of the uniform distribution.
    high (float): The upper bound of the uniform distribution.
    size (int): The number of random samples.

    Returns:
    float: The expected value of the simulated random variable.
    """
    # Simulate the random variable
    random_variable = np.random.uniform(low, high, size)

    # Calculate the expected value (mean of the samples)
    expected_value = np.mean(random_variable)

    return random_variable, expected_value

# Example usage
low = 0      # Lower bound
high = 10    # Upper bound
size = 1000  # Number of samples

random_variable, expected_value = simulate_uniform_random_variable(low, high, size)
print(f"Simulated Random Variable (first 10 samples): {random_variable[:10]}")
print(f"Expected Value: {expected_value:.2f}")
```

```python

# Q-21  Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    data1 (list or np.array): First dataset.
    data2 (list or np.array): Second dataset.
    """
    # Convert datasets to NumPy arrays
    data1 = np.array(data1)
    data2 = np.array(data2)

    # Calculate standard deviations
    std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation
    std_dev2 = np.std(data2, ddof=1)

    # Print standard deviations
    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

    # Visualization
    plt.figure(figsize=(10, 6))
    plt.hist(data1, bins=15, alpha=0.7, label=f"Dataset 1 (SD = {std_dev1:.2f})", color='blue', edgecolor='black')
    plt.hist(data2, bins=15, alpha=0.7, label=f"Dataset 2 (SD = {std_dev2:.2f})", color='orange', edgecolor='black')
    plt.title("Comparison of Standard Deviations")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

# Example Usage
dataset1 = [10, 12, 15, 18, 20, 22, 25, 27, 30, 32]
dataset2 = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

compare_standard_deviations(dataset1, dataset2)
```

```python

# Q-22 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

def calculate_range_and_iqr(mean, std_dev, size):
    """
    Generate a dataset from a normal distribution and calculate the range and interquartile range (IQR).

    Parameters:
    mean (float): Mean of the normal distribution.
    std_dev (float): Standard deviation of the normal distribution.
    size (int): Number of data points in the dataset.

    Returns:
    tuple: Range (min, max) and interquartile range (IQR).
    """
    # Generate the dataset
    data = np.random.normal(mean, std_dev, size)

    # Calculate the range
    data_range = (np.min(data), np.max(data))

    # Calculate the IQR (Q3 - Q1)
    q1 = np.percentile(data, 25)  # First quartile (25th percentile)
    q3 = np.percentile(data, 75)  # Third quartile (75th percentile)
    iqr = q3 - q1

    return data, data_range, iqr

# Example usage
mean = 50
std_dev = 10
size = 1000

data, data_range, iqr = calculate_range_and_iqr(mean, std_dev, size)
print(f"Generated Dataset (first 10 samples): {data[:10]}")
print(f"Range: {data_range[0]:.2f} to {data_range[1]:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python

# Q-23  Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Perform Z-score normalization on a dataset.

    Parameters:
    data (list or np.array): The dataset to be normalized.

    Returns:
    np.array: The Z-score normalized dataset.
    """
    # Convert data to a NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero. Cannot perform Z-score normalization.")

    # Normalize the data
    normalized_data = (data - mean) / std_dev

    return normalized_data, mean, std_dev

def visualize_transformation(original_data, normalized_data):
    """
    Visualize the transformation of original data to normalized data using histograms.

    Parameters:
    original_data (np.array): The original dataset.
    normalized_data (np.array): The Z-score normalized dataset.
    """
    plt.figure(figsize=(12, 6))

    # Plot histogram of original data
    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=20, color='skyblue', alpha=0.7, edgecolor='black')
    plt.title("Original Data Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    # Plot histogram of normalized data
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=20, color='orange', alpha=0.7, edgecolor='black')
    plt.title("Z-score Normalized Data Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

# Example usage
data = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]  # Example dataset
normalized_data, mean, std_dev = z_score_normalization(data)

print(f"Original Data Mean: {mean:.2f}")
print(f"Original Data Standard Deviation: {std_dev:.2f}")
print(f"Normalized Data (first 5 values): {normalized_data[:5]}")

visualize_transformation(data, normalized_data)
```

```python

# Q-24  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mean, std_dev, size):
    """
    Generate a dataset from a normal distribution and calculate skewness and kurtosis.

    Parameters:
    mean (float): Mean of the normal distribution.
    std_dev (float): Standard deviation of the normal distribution.
    size (int): Number of data points in the dataset.

    Returns:
    tuple: Skewness and kurtosis of the dataset.
    """
    # Generate the dataset
    data = np.random.normal(mean, std_dev, size)

    # Calculate skewness and kurtosis
    dataset_skewness = skew(data)
    dataset_kurtosis = kurtosis(data)

    return data, dataset_skewness, dataset_kurtosis

# Example usage
mean = 50
std_dev = 10
size = 1000

data, skewness, kurtosis_value = calculate_skewness_kurtosis(mean, std_dev, size)
print(f"Generated Dataset (first 10 samples): {data[:10]}")
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis_value:.4f}")
```
